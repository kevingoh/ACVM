```python
#use this for counseling event extraction
import vertexai
from vertexai.generative_models import GenerativeModel, Part, SafetySetting
from google.oauth2 import service_account
from googleapiclient.discovery import build
from pathlib import Path
import pandas as pd
import json
import os
import re
from datetime import datetime
import pandas_gbq
import pyarrow


# Initialize Vertex AI
vertexai.init(project="neuralnet-manforce", location="us-central1", credentials=credentials)

# Create credentials object with the required scopes
credentials_drive = service_account.Credentials.from_service_account_info(
    credentials_dict,
    scopes=["https://www.googleapis.com/auth/drive"]
)

# Initialize Drive service with the same credentials
drive_service = build('drive', 'v3', credentials=credentials_drive)

# Define prompt
text1 = """You are a document entity extraction specialist. Given a document, extract information into the following structure:
        {
          "details": {
            "company_name": "",
            "date_time_of_complaint": ""
          },
          "staff_details": [
            {
              "name_of_staff": "",
              "passport_number": "",
              "nationality": "",
              "problem_summary": {
                "absent": false,
                "alcohol": false,
                "bad_attitude or disobey instruction": false,
                "fighting": false,
                "runaway": false,
                "frequent_lateness_early_out": false,
                "refused_to_OT": false,
                "other_problems": null
              },
              "solution_checkboxes": {
                "briefing": false,
                "warning_letter": false,
                "stop_work": false,
                "penalty": false,
                "other_solutions": null
              }
            }
          ]
        }
        convert all cases to upper cases except in the column names and columns value in other_problems and other_solutions, and standardize the date format to use slash instead of dot.
        Make sure Go Noodle is all capital in column company name.
        Only include information found in the document. Company name is from the A. Detail section company name. synonym: alcohol=consume liquor/ too drunk, absent=absenteeism or refused to work, OT=overtime, in solution: briefing = counseling, stop work=suspend, penalty=fine.
        Capture problem summary correctly based on the checkboxes, if no checked boxes, then capture words from the statement details for each of the staff name for the problem and put in other problems column,
        For problem summary, if there is True in any columns then keep the other problems column None. there might be different problems for different staff based on the statement details. If no specific staff name mention in the statement details, then it applies to all staff. For column other problems and other solution if it is not False, then summarize the statement details/remarks instead of put True and put in the column.
        For GO NOODLE, all staff too drunk to work, means alcohol True and absent True.
        For example, in the statement details of problem summary for SMC.pdf, Rohit Nepali got complaint due to attendance & punctuality issue whereas Raju got complaint due to grooming issue, word 'grooming issue' should be stated in the other problems column for  RAJU MD RAKIB HASAN instead of true only.
        For TCRS -Seremban Mall.pdf, worker refuse to work, absent=True, then other problems=None,the solutions is briefing, worker still don't want continue work, send back hostel = stop_work, other solutions=None. For solution, list out the checkboxes as well and if not checked/filled, refer to remarks. If remarks have word 'if' infront of the solution, ignore it. for example, in George Kent.pdf, the solution remarks mention will suspend/warning letter kalau ada hal lagi, so ignore suspension and warning letter. Set values to null if not found."""

# Define generation configuration
generation_config = {
    "max_output_tokens": 8192,
    "temperature": 0,
    "top_p": 0.95,
}

# Define safety settings
safety_settings = [
    SafetySetting(category="HARM_CATEGORY_HATE_SPEECH", threshold="OFF"),
    SafetySetting(category="HARM_CATEGORY_DANGEROUS_CONTENT", threshold="OFF"),
    SafetySetting(category="HARM_CATEGORY_SEXUALLY_EXPLICIT", threshold="OFF"),
    SafetySetting(category="HARM_CATEGORY_HARASSMENT", threshold="OFF")
]

def download_file(file_id, destination):
    try:
        request = drive_service.files().get_media(fileId=file_id)
        with open(destination, 'wb') as f:
            f.write(request.execute())
        if os.path.exists(destination) and os.path.getsize(destination) > 0:
            print(f"File {destination} downloaded successfully.")
        else:
            print(f"Failed to download file {destination}.")
    except Exception as e:
        print(f"Error downloading file {file_id}: {e}")

def generate(input_file):
    document1 = Part.from_data(data=Path(input_file).read_bytes(), mime_type="application/pdf")
    model = GenerativeModel("gemini-2.0-flash-exp")
    responses = model.generate_content(
        [document1, Part.from_text(text1)],
        generation_config=generation_config,
        safety_settings=safety_settings,
        stream=True,
    )

    result = ""
    for response in responses:
        result += response.text

    # Extract JSON content within the brackets {}
    match = re.search(r'\{.*\}', result, re.DOTALL)
    if match:
        json_content = match.group(0)
        return json_content
    else:
        print("No valid JSON content found.")
        return None

def json_to_dataframe(json_output):
    company_details = json_output['details']
    staff_details = json_output['staff_details']

    # Flatten the company details
    company_flat = {
        'company_name': company_details['company_name'],
        'date_time_of_complaint': company_details['date_time_of_complaint']
    }

    # Create a list of records for the DataFrame
    records = []
    for staff in staff_details:
        problem_summary = staff.pop('problem_summary')
        solution_checkboxes = staff.pop('solution_checkboxes')
        record = {**company_flat, **staff, **problem_summary, **solution_checkboxes}
        records.append(record)

    # Create DataFrame
    df = pd.DataFrame(records)
    return df

def list_files_in_folder(folder_id):
    results = drive_service.files().list(
        q=f"'{folder_id}' in parents and mimeType='application/pdf'",
        pageSize=1000,
        fields="files(id, name, mimeType)"
    ).execute()
    return results.get('files', [])

def main():
    folder_id = '1o33LeUW49BGmeZawxSHvrJXZLqOiXzg9'  # Google Drive folder ID
    files = list_files_in_folder(folder_id)
    all_records = []

    for file in files:
        file_id = file['id']
        file_name = file['name']
        input_file = f'downloaded_{file_name}'

        download_file(file_id, input_file)

        if os.path.exists(input_file) and os.path.getsize(input_file) > 0:
            result = generate(input_file)
            if result:
                json_output = json.loads(result)
                df = json_to_dataframe(json_output)
                all_records.append(df)
                print(f"Results for {file_name}:\n{df}\n")
            else:
                print(f"Error in generating content for {file_name}.")
        else:
            print(f"File download failed or file is empty for {file_name}.")

        # Remove the downloaded file
        try:
            os.remove(input_file)
            print(f"File {input_file} removed successfully.")
        except Exception as e:
            print(f"Error removing file {input_file}: {e}")

    # Combine all records into a single DataFrame
    if all_records:
        final_df = pd.concat(all_records, ignore_index=True)
        print("Final DataFrame:")
        print(final_df)

        # Save the DataFrame to BigQuery
        table_name = f"scanned_counseling_event_df_{datetime.now().strftime('%Y%m%d')}"
        pandas_gbq.to_gbq(final_df, f'mts.{table_name}', project_id='neuralnet-manforce', if_exists='replace')
        print(f"DataFrame saved to BigQuery table: {table_name}")

if __name__ == "__main__":
    main()
```
